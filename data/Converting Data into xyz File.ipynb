```python
xyz= open('103923944.xyz', 'w')
for i in  range(len(conformations)):
    xyz.write(str(len(atomic_numbers)) + '\n')
    xyz.write(f'Frame {i+1} \n')
    for n in range(len(atomic_numbers)):
        x, y, z = conformations[i][n]
        xyz.write(str(atomic_numbers[n]) + ' ' + str(x) + ' ' + str(y) + ' ' + str(z) + '\n')
        # print(f' conformation {i+1}, atom {n+1} ')
```

```python
xyz_all= open('OCH.xyz', 'w')

atom_symbol_dict= {1: 'H', 6: 'C', 8: 'O'}

for idx in f.keys():
    atomic_number_list= list(f[idx + '/atomic_numbers'])
    energy_list= list(f[idx + '/dft_total_energy'])
    conformation_list= list(f[idx + '/conformations'])
    force_list= (list(f[idx + '/dft_total_gradient']))
    
    for i in range(len(conformation_list)):
        xyz_all.write(str(len(atomic_number_list)) + '\n')
        xyz_all.write(f'Properties=species:S:1:pos:R:3:forces:R:3 energy= {energy_list[i]} \n')
        xyz_all.write(f'Frame {i+1} \n')
        
        for n in range(len(atomic_number_list)):
            x, y, z = conformation_list[i][n]
            fx, fy, fz = force_list[i][n]
            xyz_all.write( atom_symbol_dict[atomic_number_list[n]] + ' ' + str(x) + ' ' + str(y) + ' ' + str(z) +  ' ' + str(fx) + ' ' + str(fy) + ' ' + str(fz) + '\n')
```

```python
xyz_all.close()
```
